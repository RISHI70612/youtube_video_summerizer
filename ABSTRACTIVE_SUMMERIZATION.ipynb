```python
from youtube_transcript_api import YouTubeTranscriptApi
```

```python
from IPython.display import YouTubeVideo
```

```python
video=input("Enter the link of your YouTube Video: ")
```

```python
id_video=video.split("=")[1]
print(id_video)
```

```python
YouTubeVideo(id_video)
```

```python
transcript = YouTubeTranscriptApi.get_transcript(id_video)
```

```python
doc = ""
for line in transcript:
    doc =doc+ ' ' + line['text']
```

```python
doc=[]
for line in transcript:
  if "\n" in line['text']:
    x=line['text'].replace("\n"," ")
    doc.append(x)
  else:
    doc.append(line['text'])
```

```python
paragraph=" ".join(doc)
print(paragraph)
```

## abstractive based summary

```python
from transformers import pipeline
```

```python
summarizer = pipeline('summarization')
```

```python
result=paragraph
```

```python
num_iters = int(len(result)/1000)
summarized_text = []
for i in range(0, num_iters + 1):
  start = 0
  start = i * 1000
  end = (i + 1) * 1000
  print("input text \n" + result[start:end])
  out = summarizer(result[start:end])
  out = out[0]
  out = out['summary_text']
  print("Summarized text\n"+out)
  summarized_text.append(out)
```

```python
mysummary=" ".join(summarized_text)
print(mysummary)
```
